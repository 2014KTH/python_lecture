```python
class Counter:
    def __init__(self, count, step = 1):
        self.count = count
        self.step = step
    def __str__(self):
        return '[Count (step: %d)] %d' % (self.step, self.count)
    def incr(self):
        self.count += self.step
    def __call__(self):
        self.incr()
    def __add__(self, other):
        return Counter(self.count + other, self.step)
    def __sub__(self, other):
        return Counter(self.count - other, self.step)
    def __gt__(self, y):
        return self.count > y
    def __lt__(self, y):
        return self.count < y
    def __eq__(self, y):
        return self.count == y
    def __ne__(self, y):
        return self.count != y

# 요구사항 1
c = Counter(10)
d = Counter(10, 2)
# 요구사항 2, 3
c.incr()
d.incr()
print c
print d
print
# 요구사항 3
c()
d()
print c
print d
print
# 요구사항 4
c = c + 5
d = d - 5
print c
print d
print
# 요구사항 5
print c > 10
print d > 10
print c < 10
print d < 10
print c == 17
print d != 9
```

MySet클래스는 리스트를 상속받고 있으므로, 생성자에서 리스트로 받아들인 모든 값들을 클래스 내부 리스트에 저장을 한다. 그 다음 MySet.eliminate_duplicate(self)코드가 실행되는데, eliminate_duplicate() 메소드는 @staticmethod로 선언되어있다. 이는 정적 메소드인데, 정적 메소드는 클래스 이름을 이용하여 직접 호출할 수 있는 메소드이다. 그래서 MySet.eliminate_duplicate(self)로 호출되었다. eliminate_duplicate(l) 메소드를 살펴보면, s라는 빈 리스트를 생성하고, 받아들인 리스트 l에서 for ~ in 문을 사용하여 s 리스트 안에 e라는 값이 없다면 s 리스트에 저장하도록 되어잇다. 그리고 l을 빈 리스트로 초기화하고 s에 저장된 값들을 l에 저장한다. 그러면 s = MySet([1, 2, 2, 3]) 코드가 끝난다.

그 다음 print s가 수행된다. 여기서 MySet 클래스 내부의 \_\_str\_\_ 메소드가 불린다. \_\_str\_\_ 메소드에서 result라는 변수에 중복된 값들이 제거된 리스트를 문자열로 다 저장한다. 그리고 저장된 result를 반환하면 출력이 되면서 print s 코드가 끝난다. 

그 밑의 코드 t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9]), print t도 위의 설명과 마찬가지로 수행된다.

####5. 4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    def __or__(self, other):
        a = []
        for e in self:
            a.append(e)
        for e in other:
            a.append(e)
        MySet.eliminate_duplicate(a)
        return MySet(a)

    def __and__(self, other):
        b = self | other
        c = []
        for e in b:
            if(e in self and e in other):
                c.append(e)
        return MySet(c)

    def __sub__(self, other):
        d = self | other
        f = []
        for e in d:
            if(e in self and e in other):
                f.append(e)
        for e in other:
            if(e in f):
                continue
            f.append(e)
        
        for e in f:
            d.remove(e)

        return MySet(d)
    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print s
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print t

    u = s | t
    print u

    u = s & t
    print u

    s = MySet([1,2,3])
    t = MySet([3,4,5])
    u = s - t
    print u
```

| 연산에서 우선 모든 원소를 리스트에 저장하고 그 리스트에서 중복을 제거하는 MySet.eliminate_duplicate(a) 메소드를 사용해 합집합을 만들었습니다.

& 연산에서는 우선 | 연산으로 합집합을 만들고, 그 합집합에서 두 리스트가 공통적으로 존재하는 원소만 저장해 교집합을 만들었습니다.

\- 연산에서는 우선 | 연산으로 합집합을 만들고, 그 합집합에서 두 리스트가 공통적으로 존재하는 원소 그리고 other 리스트에 있는 원소를 다른 리스트에 저장해 합집합에서 그 리스트를 빼서 차집합을 만들었습니다.

####11. [프로젝트 오일러 문제 20]
n! 이라는 표기법은 n × (n − 1) × ... × 3 × 2 × 1을 뜻합니다.

예를 들자면 10! = 10 × 9 × ... × 3 × 2 × 1 = 3628800 이 되는데,
여기서 10!의 각 자리수를 더해 보면 3 + 6 + 2 + 8 + 8 + 0 + 0 = 27 입니다.

100! 의 자리수를 모두 더하면 얼마입니까?

```python
def circulation(n):
    if n == 1:
        return 1
    return n * circulation(n-1)

save = 0
num = str(circulation(100))

for x in num:
    save += int(x)
    
print save
```

순환을 이용해서 100!의 값을 구하고 이 값을 문자열로 바꾸어서 하나의 인덱스씩 save에 더하고 save를 출력했습니다.
