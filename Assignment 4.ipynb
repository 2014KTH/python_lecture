####3. range() 함수와 유사한 frange() 함수를 다음 조건을 참고하여 만드시오.

```python
def frange(*arg):
    length = len(arg)
    if length == 1:
        save = []
        start = 0.0
        while start < arg[0]:
            save.append(float('%.1f' %start))
            start += 0.1
        return save
        
    elif length == 2:
        save = []
        start = arg[0]
        while start < arg[1]:
            save.append(float('%.1f' %start))
            start += 0.1
        return save
    
    elif length == 3:
        save = []
        start = arg[0]
        while start < arg[1]:
            save.append(float('%.1f' %start))
            start += arg[2]
        return save


print frange(0.5)
print frange(1.0, 2.0)
print frange(2.2, 4.0, 0.5)
```

매개변수를 가변인수로 받아 인자의 개수가 1개라면 range(stop)을, 인자의 개수가 2개라면 range(start, stop)을, 인자의 개수가 3개라면 range(start, stop, step)을 구현했습니다.

####4. 가변인수를 받는 함수 sum()을 다음과 같은 조건을 참고하여 구현하시오.

```python
def sum(*arg):
    total = 0
    length = len(arg)
    for i in range(length):
        total += arg[i]
    return total

print sum()
print sum(1, 2)
print sum(1,2,3,4,5)
print sum(1,5,7,2,-10)
```

####10. [프로젝트 오일러 문제 12]
1부터 n까지의 자연수를 차례로 더하여 구해진 값을 삼각수라고 합니다.

예를 들어 7번째 삼각수는 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28이 됩니다.

이런 식으로 삼각수를 구해 나가면 다음과 같습니다.

1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...
이 삼각수들의 약수를 구해봅시다.

 1: 1
 3: 1, 3
 6: 1, 2, 3, 6
10: 1, 2, 5, 10
15: 1, 3, 5, 15
21: 1, 3, 7, 21
28: 1, 2, 4, 7, 14, 28
위에서 보듯이, 5개 이상의 약수를 갖는 첫번째 삼각수는 28입니다.

그러면 500개 이상의 약수를 갖는 가장 작은 삼각수는 얼마입니까?

```python
n = 2

while(1):
    divisor = 1
    count = 0
    i = 2
    total = (n*(n+1))/2
    savetotal = total
    
    while(total >= i):
        if(total == i) :
            divisor *= 2
        if(total % i == 0):
            total /= i
            count += 1
            continue
        else:
            i += 1
            divisor *= (count+1)
            count = 0
    
    if(divisor >= 500):
        break
    else:
        n += 1

print savetotal
```

약수를 구하는 공식을 사용하여, 지수 값을 구해 divisor에 저장했습니다. divisor 즉 약수 개수가 500이 넘어간다면 삼각수를 저장한 savetotal을 출력하도록 했습니다.

####12. [프로젝트 오일러 문제 14]
양의 정수 n에 대하여, 다음과 같은 계산 과정을 반복하기로 합니다.

n → n / 2 (n이 짝수일 때)
n → 3 n + 1 (n이 홀수일 때)

13에 대하여 위의 규칙을 적용해보면 아래처럼 10번의 과정을 통해 1이 됩니다.

13 → 40 → 20 → 10 → 5 → 16 → 8 → 4 → 2 → 1
아직 증명은 되지 않았지만, 이런 과정을 거치면 어떤 수로 시작해도 마지막에는 1로 끝나리라 생각됩니다. 
(역주: 이것은 콜라츠 추측 Collatz Conjecture이라고 하며, 이런 수들을 우박수 hailstone sequence라 부르기도 합니다)

그러면, 백만(1,000,000) 이하의 수로 시작했을 때 1까지 도달하는데 가장 긴 과정을 거치는 숫자는 얼마입니까?

참고: 계산 과정 도중에는 숫자가 백만을 넘어가도 괜찮습니다.

```python
maxcount = 0
maxnum = 0

for i in range(2, 1000001):
    count = 0
    save = i
    while(i != 1):
        if(i % 2 == 0):
            i = i/2
        else:
            i = 3*i + 1
        
        count += 1
    
    if(maxcount < count):
        maxnum = save
        maxcount = count
        
print maxnum
```

백만 이하의 숫자를 선택해 그 값이 1일 때까지 짝수이면 i = i/2, 홀수이면 i = 3 * i + 1 식을 활용하고 count를 증가시켰습니다. count가 가장 큰 값을 골라 출력했습니다.

####13. [프로젝트 오일러 문제 15]
아래와 같은 2 × 2 격자의 왼쪽 위 모서리에서 출발하여 오른쪽 아래 모서리까지 도달하는 길은 모두 6가지가 있습니다 (거슬러 가지는 않기로 합니다).

그러면 20 × 20 격자에는 모두 몇 개의 경로가 있습니까?

```python
list_A = []

for i in range(21):
    list_A.append([0] * 21)

for i in range(21):
    list_A[0][i] = 1
    list_A[i][0] = 1

for i in range(1, 21):
    for j in range(1, 21):
        list_A[j][i] = list_A[j][i-1] + list_A[j-1][i]
    
print list_A[20][20]
```

모든 모서리의 리스트 원소 값을 1로 저장하고, 한 줄 한 줄 더해나가 마지막에 저장된 값이 20 x 20 격자의 모든 경로이므로 그 값을 출력했습니다.

```python
import urllib2
import string
import pickle

saveDic = {}

def fileRead(name):
    source = urllib2.urlopen(name).read()
    filename = name.replace('http://', '')
    filename1 = filename + '.html'
    f1 = open(filename1, 'w')
    f1.write(source)
    f1.close()

    i = 0
    save = ''
    save2 = []
    dic = {}

    while(i < len(source)):
        if(source[i] == '<'):
            while(source[i] != '>'):
                i += 1
            else:
                i += 1        
        else:
            save += source[i]
            i += 1        
            
    save = save.replace("//", "")
    save = save.replace("-->", "")

    stopword1 = ['a', 'and', 'are', 'as', 'at', 'be', 'but', 'by', 'for', 'if', 'in', 'into', 'is', 'it', 'no', 'not', 'of', 'on', 'or', 's', 'such', 't', 'that', 'the', 'their', 'then', 'there', 'these', 'they', 'this', 'to', 'was', 'will', 'with']
    for n in range(len(stopword1)):
        stopword1.append(stopword1[n].title())
    
    savePunc = string.punctuation
    save = save.split()
                             
    for i in range(0, len(save)):
        for punc in savePunc:
            save[i] = save[i].replace(punc, "")
        if(save[i] != ''):
            save2.append(save[i])
         
    for word in stopword1:
        count = 0
        for i in range(len(save2)):
            if save2[i] == word:
                count += 1
        for j in range(count):
            save2.remove(word)
            
    for word in save2:
        count = save2.count(word)
        dic[word] = count

    filename2 = filename + '.words.frequency'
    f2 = open(filename2, 'w')
    pickle.dump(dic, f2)
    f2.close()

    f3 = open(filename2)
    pk = pickle.load(f3)

    pkkey = pk.keys()
    pkvalue = pk.values() 

    for num in range(len(pkkey)):
        if pkkey[num] in saveDic:
            saveDic[pkkey[num]] += pkvalue[num]
        else:
            saveDic[pkkey[num]] = pkvalue[num]

    f3.close()

fileRead('http://www.nytimes.com')
fileRead('http://www.washingtonpost.com')
fileRead('http://www.google.co.kr')
fileRead('http://www.oxfordjournals.org')
fileRead('http://www.jneurosci.org')

values = saveDic.values()
values.sort(reverse = True)

result = []

for i in values:
    for j in saveDic.keys():
        if(saveDic[j] == i):
            result.append((j,i))

print result[0], result[1], result[2]
```
